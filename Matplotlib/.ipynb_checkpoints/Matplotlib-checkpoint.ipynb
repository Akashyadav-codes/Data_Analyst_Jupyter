```python
import matplotlib.pyplot as plt
import numpy as np
k=np.array([20000,30000,50000,1500000,70000,90000])
k
```

```python
plt.boxplot(k)
plt.show()
```

```python
q1=np.quantile(k,0.25)
q1
```

```python
q2=np.quantile(k,0.50)
q2
```

```python
q3=np.quantile(k,0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
upper=(q3+1.5*iqr)
upper
```

```python
lower=(q1-1.5*iqr)
lower
```

```python
k
```

```python
v=[]
for i in k:
    if ( i <lower or i> upper):
        v.append(i)
print(v)
```

```python
v2=[]
for i in k:
    if ( i<upper and i>lower):
        v2.append(i)
v2
```

```python
plt.boxplot(v2)
plt.show()
```
