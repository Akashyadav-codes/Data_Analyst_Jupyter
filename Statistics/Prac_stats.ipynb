```python
import pandas as pd

data = {
    "Student": ["A","B","C","D","E","F","G","H","I","J","K","L","M","N","O"],
    
    "Age": [18,19,20,21,20,19,22,23,21,20,19,18,22,24,23],
    
    "Math_Marks": [55,60,65,70,75,80,85,90,60,65,70,75,80,85,95],
    
    "Science_Marks": [50,55,60,65,70,75,80,85,55,60,65,70,75,80,90],
    
    "Study_Hours": [2,3,4,5,6,3,4,5,2,3,4,5,6,7,8],
    
    "Attendance": [70,75,80,85,90,78,82,88,76,79,84,86,91,93,95],
    
    "Sports_Hours": [3,2,1,2,1,3,2,1,2,3,2,1,2,1,1]
}

df = pd.DataFrame(data)

print(df)
```

```python
###   Basic Statistics
```

```python
df["Math_Marks"].mean()
```

```python
df["Math_Marks"].median()
```

```python
df["Math_Marks"].mode()
```

```python
###   Variance & Standard Deviation
```

```python
df["Math_Marks"].var()
```

```python
df["Math_Marks"].std()
```

```python
###   Quartiles
```

```python
q1=df["Math_Marks"].quantile(0.25)
q1
```

```python
q2=df["Math_Marks"].quantile(0.50)
q2
```

```python
q3=df["Math_Marks"].quantile(0.75)
q3
```

```python
iqr=q3-q1
iqr
```

```python
#    Outliers
```

```python
lower=q1-1.5*iqr
upper=q3+1.5+iqr
print(lower,upper)
```

```python
outlier=df[(df["Math_Marks"]<lower ) | (df["Math_Marks"]>upper)]
outlier
```

```python
import seaborn as sns
sns.boxplot(x=df["Math_Marks"])
```
