```python
# BREAK â€“ Movie Ticket Counter
ticket=5
person=["A","B","C","D","E","F","G"]
for people in person:
    if ticket==0:
        print(" /---- ticket sold out ------/")
        break
    print(f"Ticket given to this person {people}")
    ticket -= 1
```
