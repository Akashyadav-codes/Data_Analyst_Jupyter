```python
# Function definition
def calc_sum(a,b):  # Parameters
    c=a+b
    return c
print(calc_sum(2,3))  # function call / pass arguments
print(calc_sum(1,5))
print(calc_sum(5,6))
```

```python
# WAF TO PRINT THE ELEMENT OF A LIST IN A SINGLE LINE. ( LIST IS THE PARAMETER)
a=[1,2,3,"akash","aditya"]
def val_list(a):
    for item in a:
        print(item, end=" ")
val_list(a)
```

```python
# WAF TO FIND THE FACTORIAL OF N . ( N IS THE PARAMETER)
def fac(n):
    fac=1
    for i in range(1,n+1):
        fac=fac*i
    print(fac)
fac(5)
```

```python
RECURSION:  When a function calls itself repeatedly.
```

```python
# Print n to 1 backwards using recursion
def show(n):
    if n==0:   # Base Case: it decide recursion kb ruke
        return
    print(n)
    show(n-1)
show(5)
```

```python
# FACTOIAL USING RECURSION FUNTION
def fact(num):
    if num==0 or num==1:
        return 1
    else:
        return fact(num-1)*num
fact(5)
```
