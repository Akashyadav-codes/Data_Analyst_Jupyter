```python
Set:
1.It is collection datatypes of python.
2. Set is unordered.
3. It stored only unique element , their is no duplicasy is allowed.
4. Set support mutability ( add/remove).
5. Indexing not allowed.
```

```python
my_set = { 1,2,3,4,5,5} 
print(type(my_set))
```

```python
my_set
```

```python
# Duplicate elemnt not allowed in set.
a={1,2,3,4,5,5,6,6}
print(a)
```

```python
# Indexing not allowed
a={ 1,2,3,4}
print(a[0])
```

```python
# add()
a={ 1,2,3,4}
a.add(5)
```

```python
a
```

```python
# Update the element
a={1,2,3,4,5}
a.update([1,2,3,4,5,6])
a
```

```python
# remove() , discard()
a={1,2,3,4,5}
print(a.remove(5))
print(a.discard(4))
a
```
