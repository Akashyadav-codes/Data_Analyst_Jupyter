```python
#sum the element of nested list element in a list.
a=[1,2,[2,3,4,5],4,5,[3,4,5]]
total=0
for i in a:
    if isinstance( i , list):
        for j in i:
            total=total+j
print(total)
    
```

```python
# sum of  the didgt of each element of list

n=[12,23,12]

for i in n:
    sum=0
    for d in str(i):
        sum=sum+int(d)
    print(f"sum of the digit is {i}= {sum}")
    
```
