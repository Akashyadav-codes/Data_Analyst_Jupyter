```python
Seaborn: 
Seaborn ek Python data visualization library hai ðŸ“Š
Jo Matplotlib ke upar built hoti hai aur statistical graphs banana ka kaam easy aur attractive bana deti hai.

Thoda simple bolun to ðŸ‘‡
ðŸ‘‰ Seaborn = data ko samajhne ke liye smart + beautiful charts

Seaborn kya karta hai?

1.DataFrame (pandas) ke saath direct kaam karta hai
2.Default charts hi kaafi achhe dikhte hain
3.Statistical plots banana super easy ho jata hai

Common Seaborn plots
1.sns.lineplot() â€“ trend dekhne ke liye
2.sns.barplot() â€“ category vs value
3.sns.countplot() â€“ count nikalne ke liye
4.sns.histplot() â€“ distribution
5.sns.boxplot() â€“ outliers + spread
6.sns.heatmap() â€“ correlation / matrix
7.sns.pairplot() â€“ poora dataset ek saath
```

```python
import pandas as pd
df2= {'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings','Riders','Royals','Royals','Royals','Rcb'],
            'Rank':[1,2,2,3,3,4,1,1,2,4,1,2,1],
            'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017,2018],
            'Points':[876,789,863,673,741,812,756,788,694,701,884,690,918],
            
           }
df=pd.DataFrame(df2)
df
```

```python
df["Team"].value_counts()
```

```python
import seaborn as sns
```

```python
sns.countplot(x="Team",hue="Year",data=df)  
```

```python
sns.pairplot(df)

```

```python
s=sns.load_dataset("tips")
s
```

```python
sns.barplot(x="total_bill",y="tip",data=s)
```

```python
sns.scatterplot(x="total_bill", y ="tip",hue="sex",data=s)
```

```python
sns.lineplot(x="total_bill",y="tip",hue="sex",data=s)
```

```python
import matplotlib.pyplot as plt
```

```python
sns.relplot(x="total_bill",y="tip",hue="sex",col="day",data=s)
plt.show()
```
