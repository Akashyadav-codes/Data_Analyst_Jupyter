```python
Set:
1.It is collection datatypes of python.
2. Set is unordered.
3. It stored only unique element , their is no duplicasy is allowed.
4. Set support mutability ( add/remove).
5. Indexing not allowed.
6. union="|",intersection="&",diff="-",symmetric diff="^"
```

```python
Use case of Set:
1.✔ Duplicate remove karna
2.✔ Membership testing (in) fast
3.✔ Mathematical operations (union, intersection
```

```python
#merge the 2 set
a={1,2,3,4}
b={4,6,7,8}
print(a|b) # union
print(a&b) # intesection
print(a-b) # diff
print(a^b) # symmetric diff
```

```python
# Check two set have common element then print the value.
a={1,2,3,4}
b={2,6,7,8}
if a.isdisjoint(b):
    print(" their is no common element ")
else:
    print(" two set have  common element ")
    print(a.intersection(b))
```

```python
# it print the  symmetric diff. from two set.
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set1 ^ set2
```

```python
# find the common element from the two list.
list1=[1,3,4,5,6]
list2=[1,2,3,7,8,9]
for i in list1:
    for j in list2:
        if i==j:
            print(i)
```

```python
#FROZEN SET: Use the frozenset() constructor. Remember that frozen sets are immutable, meaning their contents cannot be changed after creation.
```

```python
# Make them in frozen set.
list=[1,2,3,4,5]
frozen_set=frozenset(list)
print(f"this is frozen set: {frozen_set}")
```
