```python
# Student Marks System (List)
# 1.Total students count karo

# 2.Highest marks nikalo

# 3.Duplicate marks find karo

# 4.Passed students (>= 40) nikaalo


marks = [78, 85, 92, 85, 60, 78]

# 1. lenth for marks show  number of studen
print(" lenth of student : ", len(marks))

# 2. find the highest marks
max = marks[0]
for m in marks:
    if m>max:
        max=m
print(" Maximum marks is :" , max)

# 3. Find the duplicate marks
duplicates = []
for m in marks:
    if marks.count(m) > 1 and m not in duplicates:
        duplicates.append(m)
print("Duplicate marks:", duplicates)

# 4. Find student who passed when greater then 40 marks
passed=[]
for i in marks:
    if i>=40:
        passed.append(i)
    else:
        print(" you are failed")
print("Passed student are:",passed)
  
```

```python
# Shopping Cart (E-commerce)
# 1.Cart me total items

# 2.Duplicate items find karo

# 3.Unique items list banao

cart=["Apple","Banana","Guava","Apple","WaterMelon"]

# 1.find number of item in cart
print(" total numbe of item:",len(cart))

# 2.Find the duplicate element in cart
dup=[]
for i in cart:
    if cart.count(i)>1 and  i not in dup:
        dup.append(i)
print("duplicate element in cart:",dup)

# 3.Find unique element and make the list
uni_list=[]
a=list(set(cart))
uni_list.append(a)
print(uni_list)
```

```python
# Mobile numbers list
numbers = [9876543210, 9123456789, 9876543210, 9988776655]

# 1. Duplicate number find karo
# 2. Unique number list banao

dup=[]
unique=[]
for i in numbers:
    if i not in unique:
        unique.append(i)
    else:
        dup.append(i)
        
print("Unique number in list is :",unique)
print(" Duplicate number in list:",dup)




 
```

```python
# Movie Ticket Booking
# To check the double times booking ticket

bookings = ["A1", "A2", "A3", "A2", "A4"]
duplicate=[]
for dup in bookings:
    if bookings.count(dup)>1 and dup not in duplicate :
        duplicate.append(dup)
print(" double times booked ticket is :",duplicate)

```
