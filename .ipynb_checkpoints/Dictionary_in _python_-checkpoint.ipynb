```python
Dictionary:
1. Dictionary denoted by {}.
2. Dictionary stores the data on key.
3. Dictionary is a mutable(changable).
4. Dictional key can't be duplicate or null.
5. Key can be any datatype.
6. Dictionary is collection of object type elements.
7. Dictionary is class of collection.
8. In dict their is no slicing because element not store at index.

```

```python
# input size from user and make the dictionary
x=int(input("enter the size of dict "))
d={}
for i in range(0,x):
    k=eval(input("enter the key"))
    v=eval(input("enter the value"))
    d.setdefault(k,v)
d
           
```
